```python
import requests
from bs4 import BeautifulSoup

# URL of the website
url = "http://quotes.toscrape.com"

# Send HTTP request
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    print("Page fetched successfully!\n")
else:
    print("Failed to fetch page")
    exit()

# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

# Extract quotes
quotes = soup.find_all("span", class_="text")

# Extract authors
authors = soup.find_all("small", class_="author")

# Extract tags
tags = soup.find_all("div", class_="tags")

# Display extracted data
for i in range(len(quotes)):
    print(f"Quote {i+1}:")
    print("Text:", quotes[i].text)
    print("Author:", authors[i].text)

    tag_list = tags[i].find_all("a", class_="tag")
    tag_names = [tag.text for tag in tag_list]

    print("Tags:", ", ".join(tag_names))
    print("-" * 50)
```

```python
import requests

url = "http://quotes.toscrape.com"
response = requests.get(url)

print(response.status_code)   # 200 means success
print(response.text[:200])    # prints first 200 characters of HTML
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")

print(type(soup))   # shows it's a BeautifulSoup object
quotes = soup.find_all("span", class_="text")

print(len(quotes))       # total quotes on page
print(quotes[0].text)    # first quote
for quote in quotes:
    print(quote.text)
    authors = soup.find_all("small", class_="author")

for author in authors:
    print(author.text)
    tags = soup.find_all("a", class_="tag")

for tag in tags[:5]:   # first 5 tags
    print(tag.text)

for i in range(len(quotes)):
    print("Quote:", quotes[i].text)
    print("Author:", authors[i].text)
    print("-" * 30)
```
